# Derivative of Matrix

## import library

````python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
````

## load image 

````python
I = img.imread('query.jpg')
````

## check the size of image

````python
n_row = I.shape[0]
n_col = I.shape[1]

print(I.shape)
````

## convert the input image into gray scale if it is color

````python
if I.shape[2] == 3:
    I = np.mean(I, axis=2)
````

## normalize input image so that the range of image is $[0, 1]$

````python
I = ((I - I.min()) / (I.max() - I.min()))
````

## generate a matrix to compute the derivative in $x$-direction

````python
Dx = np.zeros((I.shape[1], I.shape[1]))
for i in range(0, I.shape[1]-1):
    Dx[i][i] = -1
    Dx[i+1][i] = 1
Dx[I.shape[1]-1][I.shape[1]-1]= -1
````

## compute the derivative of $I$ with respect to $x$-direction

````python
Ix = I.dot(Dx)
````

```git commit -a -m "compute the derivative in x-direction"``` \
```git push origin master```

## generate a matrix to compute the derivative in $y$-direction

````python
Dy =  np.zeros((I.shape[0], I.shape[0]))
for i in range(0, I.shape[0]-1):
    Dy[i][i] = -1
    Dy[i][i+1] = 1
Dy[I.shape[0]-1][I.shape[0]-1]= -1
````

## compute the derivative of $I$ with respect to $y$-direction

````python
Iy =  Dy.dot(I)
````

```git commit -a -m "compute the derivative in y-direction"``` \
```git push origin master```

## compute $L_2^2$ of the gradient of $I$

````python
norm_gradient = np.sqrt(Ix**2 + Iy**2)
````

## define functions for the visualization

````python
def plot_image(I):
    
    plt.figure(figsize=(10,10))
    plt.imshow(I, cmap=plt.get_cmap('gray'), norm=colors.LogNorm())
    plt.title('input image')
    plt.axis('off')
        
def plot_image_derivative_x(Ix):
    
    plt.figure(figsize=(10,10))
    plt.imshow(Ix, cmap=plt.get_cmap('gray'), norm=colors.LogNorm())
    plt.title('image derivative of x')
    plt.axis('off')

    
def plot_image_derivative_y(Iy):
    
    plt.figure(figsize=(10,10))
    plt.imshow(Iy, cmap=plt.get_cmap('gray'), norm=colors.LogNorm())
    plt.title('image derivative of y')
    plt.axis('off')

    
def plot_norm_gradient(norm_gradient):

    plt.figure(figsize=(10,10))
    plt.imshow(norm_gradient, cmap=plt.get_cmap('gray'), norm=colors.LogNorm())
    plt.title('norm gradient')
    plt.axis('off')
````
